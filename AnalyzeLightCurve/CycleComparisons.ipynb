```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd

"""
IMPORTANT:
For this program to work the researcher needs to put a csv file names
combined_results_all_filter in a folder called Data. 
This folder Data will be in the same folder as this program

Expected Headers:
asteroid_id,reduced_chi2_1,reduced_chi2_2,peak_power_1,peak_power_2,
period_(hr)_1,period_(hr)_2,amplitude_1,amplitude_2,num_observations_1,
num_observations_2

OVERVIEW:
This will plot out the chi2 values, peak power values, period values, and
amplitude values on graphs to visually compare cycle 1 and cycle 2
"""

# import data from csv files
df = pd.read_csv("Data/combined_results_all_filter.csv", low_memory=False)

print(f"Asteroids: {len(df)}")
```

```python
"""
COMPARING REDUCED CHI2 VALUES
"""

plt.scatter(df["reduced_chi2_1"], df["reduced_chi2_2"])
plt.title(f"Comparing Chi2 For Cycle 1 and Cycle 2")
plt.xlabel("Chi2 Cycle 1")
plt.ylabel("Chi2 Cycle 2")
plt.grid(True)
plt.show()
```

```python
"""
COMPARING PEAK POWER VALUES
"""

plt.scatter(df["peak_power_1"], df["peak_power_2"])
plt.title(f"Comparing Peak Power For Cycle 1 and Cycle 2")
plt.xlabel("Peak Power Cycle 1")
plt.ylabel("Peak Power Cycle 2")
plt.grid(True)
plt.show()
```

```python
"""
COMPARING PERIOD VALUES
"""

plt.scatter(df["period_(hr)_1"], df["period_(hr)_2"])
plt.title(f"Comparing Period For Cycle 1 and Cycle 2")
plt.xlabel("Period (hr) Cycle 1")
plt.ylabel("Period (hr) Cycle 2")
plt.xscale("log")
plt.yscale("log")
plt.grid(True)
plt.show()
```

```python
"""
COMPARING AMPLITUDE VALUES
"""

plt.scatter(df["amplitude_1"], df["amplitude_2"])
plt.title(f"Comparing Amplitude For Cycle 1 and Cycle 2")
plt.xlabel("Amplitude Cycle 1")
plt.xlim(0,2)
plt.ylim(0,2)
plt.ylabel("Amplitude Cycle 2")
plt.grid(True)
plt.show()
```

```python
"""
Period vs Amplitude
"""

plt.scatter(df["period_(hr)_1"], df["amplitude_1"], color = "blue", label =  "Cycle 1")
plt.scatter(df["period_(hr)_2"], df["amplitude_2"], color = "red", label = "Cycle 2")
plt.title(f"Comparing Period Vs Amplitude For Cycle 1 & 2")
plt.xlabel("Period (hr)")
plt.legend()
plt.xscale("log")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
